```python
instructions = []
values = []
with open('day_02_input') as file:
    for line in file:
        splitline = line.split(' ')
        instructions.append(splitline[0])
        values.append(int(splitline[1]))
values
```

## Part 1

```python
depth = 0
hor_position = 0

for instruction, value in zip(instructions, values):
    if instruction == 'forward':
        hor_position += value
    elif instruction == 'down':
        depth += value
    elif instruction == 'up':
        depth -= value
```

```python
depth * hor_position
```

## Part 2

```python
depth = 0
aim = 0
hor_position = 0

for instruction, value in zip(instructions, values):
    if instruction == 'forward':
        hor_position += value
        depth += value * aim
    elif instruction == 'down':
        aim += value
    elif instruction == 'up':
        aim -= value
```

```python
depth * hor_position
```
