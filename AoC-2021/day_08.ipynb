```python
numberList = []
with open('day_08_input') as file:
    for line in file:
        splittedLine = line.rstrip('\n').split('|')
        twoLines = [splittedLine[0].split(), splittedLine[1].split()]
        numberList.append(twoLines)
numberList
```

```python
counter = 0
for element in numberList:
    for string in element[1]:
        if len(string) in (2, 3, 4, 7):
            counter += 1
counter
```
