```python
import numpy as np
```

```python
numberList = []
with open('day_05_input.txt') as file:
    for line in file:
        if line != '\n':
            row = line.rstrip('\n').split(' -> ')
            row = [element.split(',') for element in row]
            numberList.append(row)
```

```python
numberList
```

```python
first_array = np.zeros((1000,1000))
```

```python
for pair in numberList:
    x1 = int(pair[0][0])
    y1 = int(pair[0][1])
    x2 = int(pair[1][0])
    y2 = int(pair[1][1])
    if x1 - x2 == 0:  ## The same x coord - vertical vector
        array_with_vector = np.zeros((1000,1000))
        if y1 > y2:
            for i in range(y2, y1+1):
                array_with_vector[i, x1] = 1
        else:
            for i in range(y1, y2+1):
                array_with_vector[i, x1] = 1
        first_array += array_with_vector

    elif y1 - y2 == 0:  ## The same y coord - horizontal vector
        array_with_vector = np.zeros((1000,1000))
        if x1 > x2:
            for i in range(x2, x1+1):
                array_with_vector[y1, i] = 1
        else:
            for i in range(x1, x2+1):
                array_with_vector[y1, i] = 1
        first_array += array_with_vector
```

```python
first_array
```

```python
count = 0
for row in first_array:
    for column in row:
        if column > 1:
            count += 1
count
```

```python
second_array = np.zeros((1000,1000))
```

```python
for pair in numberList:
    array_with_vector = np.zeros((1000,1000))
    x1 = int(pair[0][0])
    y1 = int(pair[0][1])
    x2 = int(pair[1][0])
    y2 = int(pair[1][1])
    if x1 - x2 == 0:  ## The same x coord - vertical vector
        if y1 > y2:
            for i in range(y2, y1+1):
                array_with_vector[i, x1] = 1
        else:
            for i in range(y1, y2+1):
                array_with_vector[i, x1] = 1

    elif y1 - y2 == 0:  ## The same y coord - horizontal vector
        if x1 > x2:
            for i in range(x2, x1+1):
                array_with_vector[y1, i] = 1
        else:
            for i in range(x1, x2+1):
                array_with_vector[y1, i] = 1
    
    else: ## diagonal vector
        if x1 > x2:
            if y1 > y2:
                for i, j in zip(range(x2, x1+1), range(y2, y1+1)):
                    array_with_vector[j, i] = 1
            else:
                for i, j in zip(range(x2, x1+1), range(y2, y1-1, -1)):
                    array_with_vector[j, i] = 1

        else:
            if y1 > y2:
                for i, j in zip(range(x2, x1-1, -1), range(y2, y1+1)):
                    array_with_vector[j, i] = 1
            else:
                for i, j in zip(range(x1, x2+1), range(y1, y2+1)):
                    array_with_vector[j, i] = 1
    second_array += array_with_vector
```

```python
count = 0
for row in second_array:
    for column in row:
        if column > 1:
            count += 1
count
```
